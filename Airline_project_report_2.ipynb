## 1. State a Specific Goal

    A specific goal for this project is to build a predictive model that identifies the optimal time and season to purchase flight tickets using variables such as "quarter" (to capture seasonality), "fare" (average fare), "fare_low" and "fare_lg" (pricing by lowest-fare and dominant carriers), "passengers" (demand indicator), "nsmiles" (route distance), and market competition factors such as "carrier_low," "lf_ms," and "carrier_lg."

## 3. Data Preprocess

```python
# library
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
import seaborn as sns

# Import Data
flight = pd.read_csv("project2_dataset.csv")
```

#### Encode

```python
print(flight.columns)
```

```python
# Check missing value and percentage
print(flight.isnull().sum())
print((flight.isnull().mean() * 100).round(2))

# Defined ordinal columns 
ordinal_columns = ['carrier_lg', 'carrier_low']

# Defined nominal columns 
nominal_columns = ['large_ms', 'fare_lg', 'lf_ms','fare_low']

# Fill missing values with 'Unknown'
for col in ordinal_columns:
    if col in flight.columns:
        flight[col] = flight[col].fillna('Unknown')  
        
# Fill missing values with '0'
for col in nominal_columns:
    flight[col] = pd.to_numeric(flight[col], errors='coerce')
    flight[col] = flight[col].fillna(0)

# To make quarters meaningful 
flight = pd.get_dummies(flight, columns=['quarter'], dtype=int)

# Extract the dummy columns directly from the dataframe
quarter_dummies = flight[['quarter_1', 'quarter_2', 'quarter_3', 'quarter_4']]

# Create 'quarter_cat' for plotting
flight['quarter'] = (
    quarter_dummies.idxmax(axis=1)
    .str.extract(r'(\d)').astype(int)
)

# Check missing value and percentage again 
print(flight.isnull().sum())
print((flight.isnull().mean() * 100).round(2))
```

#### Merge and drop the columns 

```python
# Compare unique values between 'carrier_lg' and 'carrier_low'
col1_unique = set(flight['carrier_lg'].unique())
col2_unique = set(flight['carrier_low'].unique())
print("In column1 but not in column2:", col1_unique - col2_unique)
print("In column2 but not in column1:", col2_unique - col1_unique)

# Check if both columns have the same carrier in each row
flight['same_carrier'] = flight['carrier_lg'] == flight['carrier_low']
overlap_ratio = flight['same_carrier'].mean()
print(f"{overlap_ratio * 100:.2f}% of flights have the same carrier as largest and lowest fare.")

```

```python
# Merge the columns for make meaningful columns or make new columns that useful for analysis 

# Function to categorize 'Year' into periods
def assign_period(year):
    if 1993 <= year <= 2000:
        return '1993-2000'
    elif 2001 <= year <= 2010:
        return '2001-2010'
    elif 2011 <= year <= 2020:
        return '2011-2020'
    elif 2021 <= year <= 2024:
        return '2021-2024'
    else:
        return 'Other'

# Apply period categorization to dataset
flight['period'] = flight['Year'].apply(assign_period)


# Difference between average and lowest fare
flight['fare_gap'] = flight['fare_lg'] - flight['fare_low']

# Distance + average fare ratio (cost per mile metric)
flight['fare_per_mile'] = flight['fare_lg'] / flight['nsmiles']

print(flight['fare_gap'])
print(flight['fare_per_mile'])

# Compute the average fare per quarter and merge it back into the main dataset
avg_fare_quarter = flight.groupby('quarter')['fare'].mean().reset_index()
avg_fare_quarter.columns = ['quarter', 'avg_fare_per_quarter']
flight = flight.merge(avg_fare_quarter, on='quarter', how='left')

# Quartile Function 
def create_quartile_labels(df, column, new_col_name):
    # Calculate quartile bins while dropping NA values
    quartile_bins = pd.qcut(df[column].dropna(), 4, retbins=True, duplicates='drop')[1]

    # Create range labels based on bin edges
    quartile_labels = [
        f"{round(quartile_bins[0], 1)}$ - {round(quartile_bins[1], 1)}$",
        f"{round(quartile_bins[1], 1)}$ - {round(quartile_bins[2], 1)}$",
        f"{round(quartile_bins[2], 1)}$ - {round(quartile_bins[3], 1)}$",
        f"{round(quartile_bins[3], 1)}$ - {round(quartile_bins[4], 1)}$"
    ]

    # Assign quartile labels to the new column
    df[new_col_name] = pd.qcut(df[column], 4, labels=quartile_labels)
    return df

# Apply quartile binning to fare columns
flight = create_quartile_labels(flight, 'fare', 'fare_quartile')
flight = create_quartile_labels(flight, 'fare_lg', 'fare_lg_quartile')
flight = create_quartile_labels(flight, 'fare_low', 'fare_low_quartile')

print(flight[['fare','fare_quartile', 'fare_lg', 'fare_lg_quartile', 'fare_low', 'fare_low_quartile']].head())
```

```python
# Drop the unnecessary columns (handle the large percentage missing values)
flight.drop(['tbl', 'Geocoded_City1', 'Geocoded_City2'],axis=1, inplace=True)

print(flight.columns)
```
